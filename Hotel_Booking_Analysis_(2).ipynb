# **Chart  1 - Booking Percentage of Hotel by Name using Pie Chart**

````python
# Let's create a function which will give us bar chart of data respective with a col.

def get_count_from_column_bar(df, column_label):
    # Count the occurrences of each unique value in the specified column
    df_grpd = df[column_label].value_counts()

    # Create a new DataFrame with 'index' and 'count' columns
    df_grpd = pd.DataFrame({'index': df_grpd.index, 'count': df_grpd.values})

    return df_grpd


def plot_bar_chart_from_column(df, column_label, t1):
    # Get the counts from the specified column using the 'get_count_from_column_bar' function
    df_grpd = get_count_from_column_bar(df, column_label)

    # Create a bar chart
    fig, ax = plt.subplots(figsize=(14, 6))
    c = ['g', 'r', 'b', 'c', 'y']  # Color list for bars

    # Plot the bars
    ax.bar(df_grpd['index'], df_grpd['count'], width=0.4, align='edge', edgecolor='black', linewidth=4, color=c, linestyle=':', alpha=0.5)

    # Set chart title
    plt.title(t1, bbox={'facecolor': '0.8', 'pad': 3})

    # Add legend
    plt.legend()

    # Set y-axis label
    plt.ylabel('Count')

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=15)

    # Set x-axis label
    plt.xlabel(column_label)

    # Display the bar chart
    plt.show()
````

````python
# Chart - 1 visualization code
def get_count_from_column(df, column_label):
    # Count the occurrences of each unique value in the specified column
    df_grpd = df[column_label].value_counts()

    # Create a new DataFrame with 'index' and 'count' columns
    df_grpd = pd.DataFrame({'index': df_grpd.index, 'count': df_grpd.values})

    return df_grpd

# plot a pie chart from grouped data
def plot_pie_chart_from_column(df, column_label, t1, exp):
    # Get the counts from the specified column using the 'get_count_from_column' function
    df_grpd = get_count_from_column(df, column_label)

    # Create a pie chart
    fig, ax = plt.subplots(figsize=(14, 9))

    # Plot the pie chart
    ax.pie(df_grpd.loc[:, 'count'], labels=df_grpd.loc[:, 'index'], autopct='%1.2f%%', startangle=90, shadow=True, labeldistance=1, explode=exp)

    # Set chart title
    plt.title(t1, bbox={'facecolor': '0.8', 'pad': 3})

    # Set aspect ratio to make the pie chart circular
    ax.axis('equal')

    # Add legend
    plt.legend()

    # Display the pie chart
    plt.show()
````

````python
exp1 = [0.05, 0.05]  # Explode values for pie chart slices

# Plot a pie chart from grouped data
plot_pie_chart_from_column(hotel_booking_df, 'hotel', 'Booking percentage of Hotel by Name', exp1)
````

# **Chart 12 - Month wise Booking On a Line Chart**

````python
# Chart - 12 Visualization Code
order = ['January', 'February', 'March', 'April', 'May', 'June',
         'July', 'August', 'September', 'October', 'November', 'December']

# Filter the DataFrame based on the cancellation status
filtered_df = hotel_booking_df[hotel_booking_df['is_canceled'] == 'not canceled']

# Count the occurrences of each month and reindex based on the specified order
ordered_hotel_df = filtered_df['arrival_date_month'].value_counts().reindex(order)

# Set the figure size
plt.subplots(figsize=(10, 6))

# Set the y-axis tick positions and labels
ticks = np.arange(0, 20, 150)
labels = ["{}".format(i//1) for i in ticks]
plt.yticks(ticks, labels)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=60)

# Add annotations to each data point
for xy in zip(ordered_hotel_df.index, ordered_hotel_df.values):
    plt.annotate(text="{}".format(int(xy[1])//1), xy=xy, xytext=(0, 5), textcoords='offset points', ha='center')

# Plot the data points as a line graph
plt.plot(ordered_hotel_df.index, ordered_hotel_df.values, linewidth=4, color='r', linestyle='dotted', marker='+', markersize=20, alpha=1)

# Set the x-axis label
plt.xlabel("Months", fontdict={'fontsize': 12, 'fontweight': 5, 'color': 'Brown'})

# Set the y-axis label
plt.ylabel("Counts", fontdict={'fontsize': 12, 'fontweight': 5, 'color': 'Brown'})

# Set the title of the plot
plt.title("Month-wise Booking", fontdict={'fontsize': 20, 'fontweight': 5, 'color': 'Green'})

# Display the plot
plt.show()

````
